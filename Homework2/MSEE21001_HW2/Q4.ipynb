```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi
```

Question No 4 part b solution

```python
#here we can define the M for function
M=10

X= np.linspace(-1, 1, M).reshape([M,1])
f_x = 1-X**2
```

```python
def A_matrix_producer(X,M):
    #B=np.int64((N-1)/2)
	Pow_mat= np.linspace(M-1, 0, M).reshape([M,1]).T
	A = np.power(X,Pow_mat)
	#print(A)
	return A
```

```python
#print(X,M)
A = A_matrix_producer(X,M)
```

```python
alphas=np.dot(np.linalg.inv(np.dot(A.T,A)),np.dot(A.T,f_x))
```

```python
fx_pred = np.dot(A,alphas)
```

```python
plt.plot(X,f_x, color = 'r')
plt.scatter(X,fx_pred)
plt.legend(["actual function", "predicted function"], loc ="lower right")
plt.show()
```

Question No 4 part c solution

```python
def Q4c_func(M):
	#Setting the value of M
	

	X= np.linspace(-1, 1, M).reshape([M,1])
	f_x = 1/(1+25*X**2)

	A = A_matrix_producer(X,M)

	alphas=np.dot(np.linalg.inv(np.dot(A.T,A)),np.dot(A.T,f_x))

	fx_pred = np.dot(A,alphas)

	plt.plot(X,f_x, color = 'r')
	plt.scatter(X,fx_pred)
	plt.title(f'Plot for M={M}')
	plt.legend(["actual function", "predicted function"], loc ="lower right")
	plt.show()
```

```python
#plotting the function for M=3 to M=21 (Only odd values)
for i in range(3,22,2):
	Q4c_func(i)
```

Best choice of M

```python
#for M=9 it uses less number of samples and approximate the function well.  that's why i chose M=9 to be a better choice
Q4c_func(9)
Q4c_func(21)
```

Question No 4 part d Solution

We want to fit a polynomial of degree N 1 where N < M. Below is a python script that takes two arguments M and N. Calculates A matrics based on degree N-1 polynomial. The function to be analysed is takken from part c. The script below plots the original function (with M samples) and the predicted function (predicted on degree N-1 polynomial)

```python
def Q4d_func(M,N):
	#Setting the value of M
	

	X= np.linspace(-1, 1, M).reshape([M,1])
	f_x = 1/(1+25*X**2)

	A = A_matrix_producer(X,N)

	alphas=np.dot(np.linalg.inv(np.dot(A.T,A)),np.dot(A.T,f_x))

	fx_pred = np.dot(A,alphas)

	plt.plot(X,f_x, color = 'r')
	plt.plot(X,fx_pred)
	plt.title(f'Plot for M={M}  and N={N}')
	plt.legend(["actual function", "predicted function"], loc ="upper right")
	plt.show()
```

```python
#plotting the function for M=51 (Fixed) and changing the N
for i in range(3,22,2):
	Q4d_func(51,i)
```
